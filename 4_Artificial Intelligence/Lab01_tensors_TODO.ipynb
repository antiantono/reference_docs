There are a lot of functions and methods to initialize a tensor in pytorch, the most commons are:

* **torch.from_numpy()**    takes a numpy array and convert to a tensor
* **torch.randn**         takes samples from a Gaussian distribution (mean=0, std=1)
* **torch.rand**          takes samples from a uniform distribution [0, 1)
* **torch.ones**          creates a tensor with 1s
* **torch.zeros**         creates a tensor with 0s
* **torch.eye**           creates a Identity matrix

## Lets explore the attributes and methods

```
# lets calculate the mean of a tensor:


a = torch.randn(10,4,5)
mm = a.mean()
print(f'the mean is {mm}')


# ## TODO:  Calculate the standard deviation:
ss =  a.std()
print(f'The std is: {ss}\n')


print('Mean of the 2nd dimension across 10x4:\n',a.mean(dim=(2)))  # What it is calculating here?  What is the size of the tensor?


# ## TODO:  Calculate the standard deviation for only the first dimension of the tensor a
ss = a.std(dim = (1,2))
print(f'\nThe std is:\n {ss}  \nand has a shape of 10x1')
```

```
## Lets look closer to the mean variable:
print(mm)
# As we can see, mean is still a tensor (1-dim)
# With .item() we can obtain the value from a 1-dim tensor

print(mm.item())
print(f' The mean is :{a.mean().item()} ')
```

```
# You can also round (lower bound to next integer) or ceil (upper bound to next integer)
a = torch.randn(2,3)
print(a)

b = a.round()
print(b)

b = a.ceil()
print(b)
```

```
# You can also convert between different data types  (casting of variables)
a = torch.randn(2,3)
print(a)
print(a.dtype)
b = a.int()
print(b)
print(b.dtype )

# What did also change besides the values?
```

```
# How to obtain the proper value from a tensor?
shape = a.mean()
print(shape)

# As we can see, shape is still a tensor
```

#### Tensors data type

Data types are important because defines two aspects of the variables: Range of values and size in storage.

Lets explore what data types are available in Pytorch

```
a = torch.ones(5)
print(a.dtype)
##  Pytorch's default data type is float32 or simple floaat()
 
# Lets convert it to double()

a = a.double()
print(a.dtype)

# Convert to float 16 (half precision)
a = a.half()
print(a.dtype)


# it is possible to convert from float to integer as well.
a  = a.int()
print(a.dtype)

# and the other variants of integer
a  = a.long()
print(a.dtype)

a  = a.short()
print(a.dtype)

```

```
# It is possible to initialize a tensor with a specific data type
a = torch.empty(5, 7, dtype=torch.short)
print(a.dtype)

# Also create a tensor with a data type directly
a = torch.LongTensor(5, 7)
print(a.dtype)
```

## Operations with tensors

This section we will see different types of operations in Pytorch:

1. In-place operations. Important in Machine learning.
2. Algebraic operations on tensors: transposing, squeezing/unsqueezing, slicing, chunking and concatenating

#### In-place operations

```
# In-place operations are those methods that are applied directly on the current memory location of the variable. 
# In-place opearation are indicated '_' as in fill_(val), add_(val), etc. 

a = torch.zeros(3, 3)
a.fill_diagonal_(1)
print(a)
```

```
# with id() method it is possible to know the memory address
print(f' Memory address of tensor a: {id(a)} with value: \n {a}' )

a.multiply_(3)

print(f' Memory address of tensor a: {id(a)} with value: \n {a}' )


# Can you think in a reason about why in-place operations are important? 
# What happen if you re-run this cell?

# Mutability is the answer (and love)
```

```
# What if you do some tensor multiplications?

a = torch.randn(2,2)
print(id(a))
b = torch.randn(2,2)
print(id(b))

print(a)
print(b)
c= a*b  # matmul element_wise
print(id(c))
print(c)


# TODO: Perform this multiplication with an in-place opeation
print(a.multiply_(b))
```

Other important tensor operations: Transpositions and beyond

* Transpose:           Swaps dimension in tensor.
* Chunk:               Breaks tensor in sub-pieces.
* Concatenation:       Fuse tensors into a single-one.
* Squeeze/unsqueeze:   dimish/augment dimensions of the tensors.

```
# Transpose

# Often it is done with tensors and images. 
# Define a random image of 20x20 pixels, with format channel_first
A = torch.randn(3, 20, 20)     ## shape of tensor is [C, H, W] with first dimension is the number of channels.

# matplotlib use the channel-last format to plot images

A_ = A.transpose(0, -1)
print(f'A has a shape of: {A.shape} and A_ has a shape of: {A_.shape}')

import matplotlib.pyplot as plt
plt.imshow(A_)
```

```
# Another useful method is .view(), is similar to reshape

print(A.shape)
B = A.view(3, -1)
print(f'B has a shape: {B.shape}')

C= torch.reshape(B, (3,10,40))
print(f'C has a shape: {C.shape}')
```

```
# Chunks!

# When you want to split the tensor in sub-pieces. 
# It is necessary to determine the number of chunks and the dimension where it should be applied
# Creates a tupple with the length of the number of chunks.

A1 = torch.randn(3,2,2)
print(A1)
A_ch = torch.chunk(A1,3, dim=0)
print(f' Variable type is {type(A_ch)} and with lenght: {len(A_ch)} ')
# print(len(A_ch))
print(A_ch[0].shape)
print(f' First chunk is:\t\n {A_ch[0]} ')
print(f' Last chunk is:\t\n {A_ch[-1]} ')
```

```
#  Concatenation
A = torch.randn(4,20,20)
B = torch.randn(4,20,20)

A_B  = torch.cat([A,B], dim=0)

print(A_B.shape)

## TODO:  Create 8 chunks

A_ch = torch.chunk(A_B,8,dim=0)
print(len(A_ch))

```

Finally, we may want to add additional dimensions or remove them from our tensor.
We achieve so with [.squeeze()](https://pytorch.org/docs/stable/torch.html#torch.squeeze) or [.unsqueeze()](https://pytorch.org/docs/stable/torch.html#torch.unsqueeze).

```
# Squeeze() remove useless extra dimension in a tensor.

aa = torch.randn(10,1,3,1, 3)
print(aa.shape)

bb = aa.squeeze()
print(bb.shape)

```

```
# What if you have 3 grayscale images like below.  How you can concatenate and create a RGB image (with 3 channels).?
# Solution: Unsqueeze()

A = torch.randn(20,20)
B = torch.randn(20,20)
C = torch.randn(20,20)

plt.subplot(1,3,1)
plt.imshow(A, cmap='gray')

plt.subplot(1,3,2)
plt.imshow(B,cmap='gray')

plt.subplot(1,3,3)
plt.imshow(C,cmap='gray')
```

```

IMG_RGB  = torch.cat([A,B,C], dim=0)
print(IMG_RGB.shape)  ## Not working!!


# unsqueeze()  adds an extra-dimension to the tensor
A  = torch.unsqueeze(A, dim=0)
print(A.shape)
B  = torch.unsqueeze(B, dim=0)
C  = torch.unsqueeze(C, dim=0)
```

```
# TODO:  Concatenate A, B, C tensors and plot as an RGB image

IMG_RGB  = torch.cat([A,B,C]).squeeze().reshape([20,20,3])

print(IMG_RGB.shape)  ## Beware abouut the dimensions!!

plt.imshow( IMG_RGB )

```
